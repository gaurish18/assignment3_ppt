```python
1.def threeSumClosest(nums, target):
    nums.sort()  # Sort the array

    closestSum = float('inf')  # Initialize closest sum to a large value

    for i in range(len(nums) - 2):
        left = i + 1
        right = len(nums) - 1

        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if abs(total - target) < abs(closestSum - target):
                closestSum = total

            if total == target:
                return closestSum
            elif total < target:
                left += 1
            else:
                right -= 1

    return closestSum

# Test example
nums = [-1, 2, 1, -4]
target = 1
result = threeSumClosest(nums, target)
print(result)  # Output: 2


2.def fourSum(nums, target):
    nums.sort()  # Sort the array

    n = len(nums)
    result = []

    for a in range(n - 3):
        if a > 0 and nums[a] == nums[a - 1]:
            continue

        for b in range(a + 1, n - 2):
            if b > a + 1 and nums[b] == nums[b - 1]:
                continue

            left = b + 1
            right = n - 1

            while left < right:
                total = nums[a] + nums[b] + nums[left] + nums[right]

                if total == target:
                    result.append([nums[a], nums[b], nums[left], nums[right]])
                    left += 1
                    right -= 1

                    while left < right and nums[left] == nums[left - 1]:
                        left += 1
                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1

                elif total < target:
                    left += 1
                else:
                    right -= 1

    return result

# Test example
nums = [1, 0, -1, 0, -2, 2]
target = 0
result = fourSum(nums, target)
print(result)  # Output: [[-2, -1, 1, 2], [-2, 0, 0, 2], [-1, 0, 0, 1]]



3.def nextPermutation(nums):
    # Find the first pair of consecutive elements from the right where nums[i] < nums[i+1]
    i = len(nums) - 2
    while i >= 0 and nums[i] >= nums[i + 1]:
        i -= 1

    if i >= 0:
        # Find the smallest element greater than nums[i] in the suffix
        j = len(nums) - 1
        while j > i and nums[j] <= nums[i]:
            j -= 1

        # Swap nums[i] with nums[j]
        nums[i], nums[j] = nums[j], nums[i]

    # Reverse the suffix
    left = i + 1
    right = len(nums) - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1

    return nums

# Test example
nums = [1, 2, 3]
result = nextPermutation(nums)
print(result)  # Output: [1, 3, 2]



4.def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left

# Test example
nums = [1, 3, 5, 6]
target = 5
result = searchInsert(nums, target)
print(result)  # Output: 2




5.def plusOne(digits):
    n = len(digits)
    for i in range(n - 1, -1, -1):
        if digits[i] == 9:
            digits[i] = 0
        else:
            digits[i] += 1
            return digits

    # If the loop completes without returning, it means all digits were 9s
    return [1] + digits

# Test example
digits = [1, 2, 3]
result = plusOne(digits)
print(result)  # Output: [1, 2, 4]




6.def singleNumber(nums):
    result = 0
    for num in nums:
        result ^= num
    return result

# Test example
nums = [2, 2, 1]
result = singleNumber(nums)
print(result)  # Output: 1




7.def findMissingRanges(nums, lower, upper):
    result = []
    start = lower

    for num in nums + [upper + 1]:
        if num == start:
            start += 1
        elif num > start:
            end = num - 1
            result.append([start, end])
            start = num + 1

    return result

# Test example
nums = [0, 1, 3, 50, 75]
lower = 0
upper = 99
result = findMissingRanges(nums, lower, upper)
print(result)  # Output: [[2, 2], [4, 49], [51, 74], [76, 99]]




8.def canAttendMeetings(intervals):
    intervals.sort(key=lambda x: x[0])  # Sort intervals based on start time

    for i in range(1, len(intervals)):
        if intervals[i][0] < intervals[i-1][1]:
            return False

    return True

# Test example
intervals = [[0, 30], [5, 10], [15, 20]]
result = canAttendMeetings(intervals)
print(result)  # Output: False
```
